```python
import json
import boto3
import datetime
import tweepy

def default(o):
    if isinstance(o, (datetime.date, datetime.datetime)):
        return o.isoformat()
      
secret_name = "twitter"
region_name = "us-east-1"

# Create a Secrets Manager client
session = boto3.session.Session()
client = session.client(
    service_name='secretsmanager',
    region_name=region_name
)

get_secret_value_response = client.get_secret_value(
        SecretId=secret_name)

credenciais = json.loads(get_secret_value_response.get("SecretString"))

CONSUMER_KEY = credenciais.get("CONSUMER_KEY")
CONSUMER_SECRET = credenciais.get("CONSUMER_SECRET")
ACCESS_TOKEN = credenciais.get("ACCESS_TOKEN")
ACCESS_TOKEN_SECRET = credenciais.get("ACCESS_TOKEN_SECRET")

# Authenticate to Twitter
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

# Create API object
api = tweepy.API(auth)

# Create a tweet
queries = ["#EleNao", "Bolsonaro"]
lista_tweets = dict()
for query in queries:
    lista_tweets[query] = dict()
    latest_tweets = api.search(q=query, lang="pt-br")
    for tweet in latest_tweets:
        lista_tweets[query][tweet._json["id"]] = tweet._json
```
